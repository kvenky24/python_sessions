# Lambda

**One Arguement**

```python
def add(x):
    return(x+10)
add(20)
```

```python
add=lambda x:x+10
add(20)
```

```python
# def sqaure(x):
#     return(x*x)
# sqaure(5)

sqaure=lambda x:x*x
sqaure(5)
```

```python
cube=lambda a:a*a*a
cube(3)
```

**Two+ Arguements**

```python
def add(a,b):
    return(a+b)
add(2,3)
```

```python
add=lambda a,b:a+b
add(2,3)
```

```python
def avg(a,b,c):
    return((a+b+c)/3)
avg(4,6,8)
```

```python
avg=lambda a,b,c:(a+b+c)/3
avg(4,6,8)
```

```python
avg=lambda a,b,c=5:(a+b+c)/3
avg(4,6)
```

```python
avg=lambda a,b,c=51:round((a+b+c)/3,2)
avg(4,6)
```

$If-Else:$

```python
def greater(n,m):
    if n>m:
        return(n)
    else:
        return(m)
greater(200,300)
```

```python
l1=[]
def greater(n,m):
    if n>m:
        l1.append(n)
    else:
        l1.append(m)
greater(200,300)
```

```python
l1
```

```python
greater=lambda n,m:(n if n>m else m)
```

```python
greater(2,3)
```
