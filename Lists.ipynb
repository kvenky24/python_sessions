$List -Comprehension:$

```python
list4=[1,2,3,4,5]
output=[]
for i in list4:
    output.append(i*i)
print('output:',output)

output1=[i*i for i in list4]
print('output1:',output1)
```

```python
list1=["Who is PM of India?","What is the Capital of India?","Who is the President of India?"]
list2=["Modi","Delhi","Draupadi"]
marks=0
correct_answer=0
for i in range(len(list1)):
    print(list1[i])
    answer=input("Enter Your Answer: ")
    if answer.lower()==list2[i]:
        marks+=1
print(correct_answer)
```

```python
Question=["Who is PM of India?","What is the Capital of India?","Who is the President of India?"]
Answer=["Modi","Delhi","Draupadi"]
[marks:=marks+1 for i in range(len(Question)) if input(Question[i]).capitalize()==Answer[i]]
print(marks)
```
