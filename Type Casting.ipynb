### Type Casting
- Int - Other Data Types
- Float - Other Data Types

**Integer to other data types**

```python
number1=100
number1
```

```python
type(number1)
```

```python
float_number1=float(number1)
```

```python
float_number1
```

```python
string_number1=str(number1)
```

```python
string_number1
```

```python
boolean_number=bool(number1)
```

```python
boolean_number
```

```python
type(boolean_number)
```

```python
complex_number=complex(number1)
```

```python
complex_number
```

**Float to other Data types**

```python
number2=100.5
number2
```

```python
int_number=int(number2)
int_number
```

```python
str_number=str(number2)
str_number
```

```python
bool_number=bool(number2)
bool_number
```

```python
complex_number=complex(number2)
complex_number
```
