```python
def game():
    import random
    random_num=random.randint(1,10)
    chances=eval(input("Enter a number of chances you want: "))
    for i in range(chances):
        user_num=eval(input("Enter a number between 1 and 10:"))
        if user_num==random_num:
            print("You Won")
            break
        else:
            print("You Lost,Because Random num is {}".format(random_num))
            print("No. of chances left is:",chances-1-i)
```

```python
account_bal=200
ewal_bal=5000
for i in range(3):
    j=eval(input("Enter amount you need to withdraw from e-wallet: "))
    account_bal=account_bal+j
    ewal_bal=ewal_bal-j
    if account_bal>=1000:
        print("Now,You are ready to play the Game.")
        print("Your E-wallett Balance is {} & account_bal is {}".format(ewal_bal,account_bal))
        break
    else:
        print("You still do not have enough money to play the Game.")
        print("Your E-wallett Balance is {} & account_bal is {}. You need {}".format(ewal_bal,account_bal,1000-account_bal))
```
